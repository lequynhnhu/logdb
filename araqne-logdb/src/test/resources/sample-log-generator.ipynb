```python
import random
import datetime
```

```python
sample_format = "{date} {rid} {cat1} {cat2} {val1} {val2}"
```

```python
row_template = { 'date': None, 'cat1': None, 'cat2': None, 'val1': None, 'val2': None, }
```

```python
dists = {
    'A': lambda : random.normalvariate(10, 3),
    'B': lambda : random.normalvariate(20, 3),
    'C': lambda : random.normalvariate(30, 3)
    }
```

```python
row_size = 3600
```

```python
date_base = datetime.datetime(2013, 10, 1, 0, 0, 0)
```

```python
filename = r'SummaryLoggerTest-Sample1.txt'
```

```python
import random
from itertools import takewhile

def accumulate(iterator):
    """Returns a cumulative sum of the elements.
    accumulate([1, 2, 3, 4, 5]) --> 1 3 6 10 15"""
    current = 0
    for value in iterator:
        current += value
        yield current

def weighted_choice(weights, objects):
    """Return a random item from objects, with the weighting defined by weights 
    (which must sum to 1)."""
    limit = random.randint(0, sum(weights))
    return objects[sum(takewhile(bool, (value < limit for value in accumulate(weights))))]

def get_date(date_base, interval, count):
    return date_base + datetime.timedelta(seconds = interval * count)
    
```

```python
f = open(filename, 'w')

for rid in xrange(row_size):
    row = dict(row_template)
    row['rid'] = str(rid)
    row['date'] = str(get_date(date_base, 1, rid))
    row['cat1'] = weighted_choice([1, 2, 3], ['A', 'B', 'C'])
    row['cat2'] = weighted_choice([1, 1], ['X', 'Y'])
    row['val1'] = int(dists[row['cat1']]())
    row['val2'] = int(dists[row['cat1']]() + 100)
    f.write(sample_format.format(**row) + '\n')
    #print row
    print sample_format.format(**row)
    
f.close()
    
```
